```python
import os
import json
import pandas as pd
import git
#extract aggregate transaction data from json file

def agg_trans_data():
    path1='pulse/data/aggregated/transaction/country/india/state/'
    agg_trans_state_list=os.listdir(path1)

    tran_clm={'state':[],'year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for state in agg_trans_state_list:
        p_state=path1 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for i in d['data']['transactionData']:
                    name=i['name']
                    count=i['paymentInstruments'][0]['count']
                    amount=i['paymentInstruments'][0]['amount']
                    tran_clm['Transaction_type'].append(name)
                    tran_clm['Transaction_count'].append(count)
                    tran_clm['Transaction_amount'].append(amount)
                    tran_clm['state'].append(state)
                    tran_clm['year'].append(year)
                    tran_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(tran_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')

    return(df)
```

```python
#extract map transaction data from json file

def map_trans_data():
    path3="pulse/data/map/transaction/hover/country/india/state/"
    map_trans_state_list=os.listdir(path3)

    tran_clm={'state':[],'District':[],'year':[],'Quater':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for state in map_trans_state_list:
        p_state=path3 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for i in d['data']['hoverDataList']:
                    name=i['name']
                    count=i['metric'][0]['count']
                    amount=i['metric'][0]['amount']
                    tran_clm['District'].append(name)
                    tran_clm['Transaction_count'].append(count)
                    tran_clm['Transaction_amount'].append(amount)
                    tran_clm['state'].append(state)
                    tran_clm['year'].append(year)
                    tran_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(tran_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    df['District']=df['District'].str.replace('district','').str.title()

    return(df)
```

```python
def map_user_data():
    path4="pulse/data/map/user/hover/country/india/state/"
    map_user_state_list=os.listdir(path4)

    user_clm={'state':[],'District':[],'year':[],'Quater':[],'Registered_Users':[],'App_Opens':[]}

    for state in map_user_state_list:
        p_state=path4 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for district,data_dist in d['data']['hoverData'].items():
                    registeredUsers=data_dist['registeredUsers']
                    appopens=data_dist['appOpens']
                    user_clm['District'].append(district)
                    user_clm['Registered_Users'].append(registeredUsers)
                    user_clm['App_Opens'].append(appopens)
                    user_clm['state'].append(state)
                    user_clm['year'].append(year)
                    user_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(user_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    df['District']=df['District'].str.replace('district','').str.title()
    return(df)
```

```python
#extract top transaction data from json file

def top_trans_data():
    path5="pulse/data/top/transaction/country/india/state/"
    top_trans_state_list=os.listdir(path5)

    tran_clm={'state':[],'Pincode':[],'year':[],'Quater':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for state in top_trans_state_list:
        p_state=path5 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for i in d['data']['pincodes']:
                    pincode=i['entityName']
                    count=i['metric']['count']
                    amount=i['metric']['amount']
                    tran_clm['Pincode'].append(pincode)
                    tran_clm['Transaction_count'].append(count)
                    tran_clm['Transaction_amount'].append(amount)
                    tran_clm['state'].append(state)
                    tran_clm['year'].append(year)
                    tran_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(tran_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')

    return(df)
```

```python
#extract top user data from json file

def top_user_data():
    path6="pulse/data/top/user/country/india/state/"
    map_user_state_list=os.listdir(path6)

    user_clm={'state':[],'Pincode':[],'year':[],'Quater':[],'Registered_Users':[]}

    for state in map_user_state_list:
        p_state=path6 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for i in d['data']['pincodes']:
                    pincode=i['name']
                    registeredUsers=i['registeredUsers']
                    user_clm['Pincode'].append(pincode)
                    user_clm['Registered_Users'].append(registeredUsers)
                    user_clm['state'].append(state)
                    user_clm['year'].append(year)
                    user_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(user_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    return(df)
```

```python
import sqlite3

# Create a SQLite database (this will create a file named 'phonepe_data.db' in your Colab environment)
conn = sqlite3.connect('phonepe_data.db')

# Convert the DataFrames to SQLite tables
# Assuming agg_trans_df, agg_user_df, map_trans_df, map_user_df, top_trans_df, top_user_df are the names of your dataframes.
# If you used different names, please adjust these accordingly.

# Call functions to get the dataframes
agg_trans_df = agg_trans_data()
agg_user_df = agg_user_data()
map_trans_df = map_trans_data()
map_user_df = map_user_data()
top_trans_df = top_trans_data()
top_user_df = top_user_data()


# Store each DataFrame into a separate table in the SQLite database
agg_trans_df.to_sql('aggregated_transactions', conn, if_exists='replace', index=False);
agg_user_df.to_sql('aggregated_users', conn, if_exists='replace', index=False);
map_trans_df.to_sql('map_transactions', conn, if_exists='replace', index=False);
map_user_df.to_sql('map_users', conn, if_exists='replace', index=False);
top_trans_df.to_sql('top_transactions', conn, if_exists='replace', index=False);
top_user_df.to_sql('top_users', conn, if_exists='replace', index=False);

# Commit changes to the database
conn.commit()

print("All data has been successfully stored in 'phonepe_data.db' SQLite database.")

# Close the connection (important after you're done writing data)
conn.close()
```
